### Morphological Transformation


Morphological operations are a set of image processing techniques used for analyzing and processing geometric structures within images. These operations are based on the shape and structure of objects within an image rather than their numerical pixel values.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/morph.png', 0)
plt.imshow(image,"gray")
plt.axis("off")
```

#### Dilation

Dilation is a morphological operation that expands the boundaries of an object in an image. This is done by convolving the image with a structuring element (kernel), which determines the size and shape of the dilation. 

```python
kernel = np.ones((7,7), np.uint8)
```

```python
dilatedImage = cv2.dilate(image, kernel, iterations=1)
plt.imshow(dilatedImage,"gray")
plt.axis("off")
plt.imsave("Dilation.png",dilatedImage,cmap="gray")
```

#### Erosion

Erosion is a morphological operation that shrinks the boundaries of an object in an image. This is achieved by convolving the image with a structuring element (kernel) that defines the size and shape of the erosion.

```python
erodedImage = cv2.erode(image, kernel, iterations=1)
plt.imshow(erodedImage,"gray")
plt.axis("off")
plt.imsave("Erosion.png",erodedImage,cmap="gray")
```

#### Closing

Closing is a morphological operation that combines dilation followed by erosion. It is commonly used to fill small gaps and holes in objects, as well as to connect broken lines or separate objects that are close to each other. 

```python
# Closing (dilation followed by erosion)
closingImage = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
plt.imshow(closingImage,"gray")
plt.axis("off")
plt.imsave("Closing.png",closingImage,cmap="gray")
```

#### Opening
Opening is a morphological operation that combines erosion followed by dilation. It is primarily used to remove noise and small objects from images while preserving the shape and size of larger objects.

```python
# Opening (erosion followed by dilation)
openingImage = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
plt.imshow(openingImage,"gray")
plt.axis("off")
plt.imsave("Opening.png",openingImage,cmap="gray")
```

#### Gradient

The gradient is a morphological operation used in image processing to highlight the edges and boundaries of objects within an image. Specifically, the gradient is computed as the difference between the dilation and erosion of the input image.

```python
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient,"gray")
plt.axis("off")
plt.imsave("Gradient_morph.png",gradient,cmap="gray")
```

#### Skeleton

In image processing, the skeleton (morphological skeleton or medial axis) of an object is a thin representation of its structure, capturing its essential shape and topology. The skeleton is obtained through a series of morphological operations applied to the binary representation of the object in the image.

```python
skeleton = cv2.ximgproc.thinning(image)
plt.imshow(skeleton,"gray")
plt.axis("off")
plt.imsave("Skeleton.png",skeleton,cmap="gray")
```
